```python
import pandas as pd
import numpy as np
from numpy import nan as NA
from pandas import Series, DataFrame
import matplotlib
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Read the final csv for part 2
customer_2 = pd.read_csv("customer_final.csv")
customer_2.head(5)
```

```python
# Create the menu with 5 different choices
def myMenu(df):
    quit = False
    while quit == False:
        print("\nLet's see the menu!")
        print("1. See the first 15 rows of your dataframe")
        print("2. See the median number for each quantitative column")
        print("3. Get the histogram distribution charts for each quantitative column")
        print("4. Get the pie charts for each qualitative column")
        print("5. Quit")
        menu_choice = input("What can I get you today?")
        try:
            menu_choice = int(menu_choice)
        except:
            print("Oops! You enter a text or a symbol. Enter a whole number for 1 to 5 only")
            continue
        if menu_choice not in [1, 2, 3, 4, 5]:
            print("Oops! Please enter 1, 2, 3, 4, 5 only")
        else:
            if menu_choice == 1:
                print("\nLet's see the first 15 rows of your dataframe!")
                print(df.head(15))
            if menu_choice == 2:
                print("\n Fine the median for each quantitative column")
                for c in df.select_dtypes('number').columns:
                    print(f"{c}: {customer_2[c].median():.2f}")
            if menu_choice == 3:
                print("\nLet's see the histogram distribution charts for each quantitative column")
                for i, c in enumerate(df.select_dtypes('number').columns):
                    plt.figure(i)
                    sns.displot(df[c])
                    plt.show()
            if menu_choice == 4:
                print("\nLet's see the pie charts for each qualitative column")
                for i, c in enumerate(df.select_dtypes('object').columns):
                    plt.figure(i)
                    df[c].value_counts().plot.pie(subplots = True, figsize = (2,2))
                    plt.show()
            if menu_choice == 5:
                quit = True
myMenu(customer_2)
```
